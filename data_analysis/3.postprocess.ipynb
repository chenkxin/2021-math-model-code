```python
def remove_outlier(df_in, col_name):
    df_in[col_name] = df_in[col_name].apply(pd.to_numeric, errors='coerce')
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

def remove_outlier_from_multi_cols(df_in, col_names):
    for c in col_names:
        df_in = remove_outlier(df_in, c)
    return df_in

for place in PLACES:
    frame = {str(i):None for i in range(3)}
    for i in range(3):
        odf = dfs[place][str(i)]
        df = dfs[place][str(i)]
        df = remove_outlier_from_multi_cols(df, find_pollution_columns(df))

        fig, axs = plt.subplots(6, 2, figsize=(12,20))
        x = df.columns[2] if i == 0 else df.columns[1]
        for j, p in enumerate(find_pollution_columns(df)):
            odf.plot(x=x, y=p,  ax=axs[j][0])
            df.plot(x=x, y=p,  ax=axs[j][1], c="green")

        fig.tight_layout()
        plt.savefig(f"results/异常值处理{place}_{i}."+FORMAT, format=FORMAT)
        plt.close()
        
        frame[str(i)] = df
    
    write_excel(f"{place}", frame)
```
